```python
def important_features(dataset):
    data=dataset.copy()
    for i in range(0, dataset.shape[0]):
        data["imp"]=data["movie_title"]+' '+data["director_name"]+" " +data["genres"]
    return data
```

## COSINE SIMILARITY

```python
##Importing vectoriser
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
vec=TfidfVectorizer()
```

```python
vecs=vec.fit_transform(rec_data["plot_keywords"].apply(lambda x: np.str_(x)))
```

```python
vecs.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vecs) 
```

```python
sim.shape ##Columns represnt movie-ids and not rows
```

```python
# To remove the "\xa0" character at the end of each movie title string
rec_data["movie_title"] = rec_data["movie_title"].apply(lambda x: x.rstrip("\xa0"))
```

```python
def recommend(title):
    movie_id =rec_data[rec_data.movie_title==title]["ids"].values[0]
    scores=list(enumerate(sim[movie_id]))
    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
    sorted_scores=sorted_scores[1:]
    movies=[rec_data[movies[0]==rec_data["ids"]]["movie_title"].values[0] for movies in sorted_scores]
    return movies
```

```python
def recommend_ten(movie_list):
    first_ten=[]
    count=0
    for movie in movie_list:
        if count > 9:
            break
        count+=1
        first_ten.append(movie)
    return first_ten
```

```python
lst=recommend("Zathura: A Space Adventure")
m=recommend_ten(lst)
```

## KNN MODEL IMPLEMENTATION

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Fit and transform the plot descriptions to TF-IDF vectors
plot_vectors = tfidf.fit_transform(rec_data['imp'].fillna(''))

# Build a KNN model using cosine similarity
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(plot_vectors)

def KNN_recommend(movie_title, n=10):
    # Get the index of the movie with the given title
    movie_idx = rec_data[rec_data['movie_title'] == movie_title].index[0]

    # Get the top n recommendations based on cosine similarity
    distances, indices = knn.kneighbors(plot_vectors[movie_idx], n_neighbors=n+1)
    # note that the first recommendation will be the movie itself, so we ignore it
    indices = indices.flatten()[1:]

    # Return the recommended movie titles
    return list(rec_data.loc[indices, 'movie_title'])
```

```python
KNN_recommend("The Children of Huang Shi")
```

## SYSTEM TESTING

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

# Load the dataset
# data = pd.read_csv('./Data/movie_metadata.csv', encoding='latin1')
data = rec_data.copy()
data = data[data['genres'] != '']

# Create a new column with the genre name only
data['genre'] = data['genres'].apply(lambda x: x.split('|')[0])

# Filter out the movies with no plot keywords specified
data = data[data['imp'].notna()]


# Split the data into train and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Create the count vectorizer and fit it to the train data
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_data['imp'])
X_test = vectorizer.transform(test_data['imp'])
y_train = train_data['genre']
y_test = test_data['genre']

# Create the multinomial naive Bayes classifier and fit it to the train data
clf = MultinomialNB()
clf.fit(X_train, y_train)

# Make predictions on the test data and calculate precision, recall, f1-score, and accuracy
y_pred = clf.predict(X_test)
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
accuracy = accuracy_score(y_test, y_pred)

# Print the results
print('Precision: {:.2f}'.format(precision))
print('Recall: {:.2f}'.format(recall))
print('F1-score: {:.2f}'.format(f1_score))
print('Accuracy: {:.2f}'.format(accuracy))
```

```python
# Define the labels and data for the bar chart
labels = ['Precision', 'Recall', 'F1-score', 'Accuracy']
data = [precision, recall, f1_score, accuracy]

# Create the bar chart using Seaborn
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=labels, y=data, palette='viridis')

# Set the chart title and axis labels
plt.title('Model Performance')
plt.xlabel('Metrics')
plt.ylabel('Scores')

# Display the chart
plt.show()
```
